```python
import spherometer as sph
import matplotlib.pyplot as plt
import sympy as sy
import numpy as np

plt.style.use('dark_background')
sy.init_printing()
```

```python
inst = sph.Spherometer.load()

s = np.linspace(0.01, 0.18, 1000)
fl = inst.func_f()(s)
f_ratio = inst.func_f_ratio()(s, 6)

fig, ax = plt.subplots(figsize=(10, 10))
ax.plot(s, fl, '.w')

ax2 = ax.twinx()

ax.set_xlim(0, 0.18)
ax.set_ylim(0, 100)
ax.set_xlabel('Sagitta [in]')
ax.set_ylabel('Focal Length [in]')
ax2.set_ylim(0, 100/6)
ax2.tick_params(axis='y', labelcolor='r')
ax2.set_ylabel('Focal Ratio', color='r')
ax.axvline(0.0235)
ax.axvline(0.0270)

plt.show()
```
